```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests


```

```python
url=("https://www.indexmundi.com/commodities/?commodity=crude-oil-brent&months=300")
page=requests.get(url)
print(page)
```

```python
beauty=BeautifulSoup(page.text,'html') # containg the HTML text of the webpage
print(beauty)
```

```python
print(beauty.prettify())
```

```python
oil=beauty.find_all("div")
#print(oil)
```

```python
title=beauty.find("div",class_='<th scope="col">Month</th><th scope="col">Price</th><th scope="col">Change</th>')
```

```python
world_titles=beauty.find_all('th')  # Finding all the title names
world_titles
```

```python
#for title in world_titles:
 # print(title.text)

world_beauty_title=[title.text for title in world_titles]
world_beauty_title
```

```python
df=pd.DataFrame(columns=world_beauty_title)
df
```

```python
column_data=beauty.find_all('tr')
#print(column_data)
for row in column_data:
    row_data=row.find_all('td')
    #print(individual_row_data)
    if len(row_data) == len(df.columns):
        individual_row_data=[data.text.strip() for data in row_data]
        length=len(df)
        df.loc[length]=individual_row_data
print(df)
```
